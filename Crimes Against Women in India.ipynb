# Crimes Against Women in India (2001-2021)



####  In this presentation, we will explore data on crimes against women in India over the past 20 years. We will look at different types of crimes, such as rape, domestic violence, and dowry deaths, and see how these crimes have changed over time and across different states.

##### Column Description


* Unnamed: 0: Index column (can be ignored).
* State: Name of the state or union territory.
* Year: Year of the recorded data.
* Rape: Number of rape cases.
* K&A: Kidnapping and abduction cases.
* DD: Dowry death cases.
* AoW: Assault on women cases.
* AoM: Insult to modesty cases.
* DV: Domestic violence cases (by husbands or relatives).
* WT: Women trafficking cases.

##### This dataset contains a comprehensive view of various crimes against women across different states in India over multiple years.

### Cleaning and standardizing the 'State' column by stripping spaces and converting to lowercase

```python
df['State']=df['State'].str.strip().str.lower()
df['State'].unique()
```

## Extracting Total Crimes Column

```python
df['Total Crimes'] = df[['Rape', 'Kidnapping and Abduction', 'Dowry Deaths', 
                         'Assault against Women', 'Insult to Modesty of Women', 
                         'Domestic Violence', 'Women Trafficking']].fillna(0).sum(axis=1)
```

```python
df
```

## Trends in Total Crimes Against Women in India (2001-2021)

```python
plt.figure(figsize=(10,6))
sns.lineplot(x=df['Year'],y=df['Total Crimes'], marker='o',estimator='sum')
plt.title('Crimes Against Women (2001-2021)')
plt.xlabel('Year')
plt.ylabel('Total Crimes')
plt.grid(True)
plt.show()
```

### Objective:
* To analyze and visualize the changes in the total number of crimes against women in India over the years from 2001 to 2021
* To identify any significant increases or decreases in crime rates

* #### The plot shows a general increase in crimes against women from 2001 to 2021, with noticeable fluctuations and a significant rise in recent years.

* #### Increasing Trend: The National Crime Records Bureau (NCRB) reported a continuous rise in crimes against women, with incidents of rape increasing from approximately 18,000 cases in 2001 to over 32,000 in 2021.

## Top 10 States with the Highest Total Crimes Against Women in India (2001-2021)

```python
statewise_crimes = df.groupby('State')['Total Crimes'].sum().reset_index().sort_values(by="Total Crimes",ascending=False).reset_index(drop=True).head(10)
statewise_crimes
```

```python
plt.figure(figsize=(10,6))
sns.barplot(x='State', y='Total Crimes', data=statewise_crimes, palette='Reds_d')
plt.title('Top 10 States with Highest Crime Rates')
plt.xlabel('States')
plt.ylabel('Total Crimes')
plt.xticks(rotation=90)
plt.show()
```

### Objective:
* To visualize and compare the total number of crimes against women across different states in India, focusing on the top 10 states with the highest crime rates.
* This analysis aims to highlight regions that require urgent attention and targeted interventions to address the prevalence of violence against women.

* #### Uttar Pradesh has the highest crime rate among the top 10 states in India, with a significant lead over the others.

* #### Uttar Pradesh accounted for around 14% of total reported rapes in 2021, highlighting the need for focused interventions in these regions.




## Comparison of Crimes Against Women Across Top 10 States

```python
statewise_crimetypesum = df.groupby('State')[['Total Crimes','Rape', 'Kidnapping and Abduction', 'Dowry Deaths', 
                                'Assault against Women', 'Insult to Modesty of Women', 
                                'Domestic Violence', 'Women Trafficking']].sum().reset_index().sort_values(by="Total Crimes",ascending=False).reset_index(drop=True).head(10)
statewise_crimetypesum
```

```python
melted = statewise_crimetypesum.melt(id_vars='State', 
                                        value_vars=['Rape', 'Kidnapping and Abduction', 'Dowry Deaths', 
                                                    'Assault against Women', 'Insult to Modesty of Women', 
                                                    'Domestic Violence', 'Women Trafficking'],
                                        var_name='Crime Type', 
                                        value_name='Number of Crimes')
melted
```

```python
plt.figure(figsize=(14, 6))
sns.barplot(x='State', y='Number of Crimes', hue='Crime Type', data=melted, palette='Paired')
plt.title('Crime Categories Across Top 10 States')
plt.xlabel('States')
plt.ylabel('Number of Crimes')
plt.xticks(rotation=45)
plt.show()
```

* ####  Here are the top 10 states witnessing the highest number of domestic violence cases.

## Analyzing the Distribution of Different Crimes Over Time



```python
crimetypesum = df.groupby('Year')[['Rape', 'Kidnapping and Abduction', 'Dowry Deaths', 
                                'Assault against Women', 'Insult to Modesty of Women', 
                                'Domestic Violence', 'Women Trafficking']].sum()
crimetypesum
```

```python
crimetypesum.plot(kind='bar', stacked=True, figsize=(12, 6), colormap='Set1')
plt.title('Distribution of Different Crimes Against Women Over Time (2001-2021)')
plt.xlabel('Year')
plt.ylabel('Total Crimes')
plt.show()
```

* #### It shows a steady rise in crimes against women in India from 2001 to 2021, with domestic violence and sexual assault being the most common.

## Crime Type Distribution in India (2021)

```python
year = 2021  
df_year = df[df['Year'] == year]
crime_data2021 = df_year[['Rape', 'Kidnapping and Abduction', 'Dowry Deaths', 
                      'Assault against Women', 'Insult to Modesty of Women', 
                      'Domestic Violence', 'Women Trafficking']].sum()

crime_data2021
```

```python
plt.figure(figsize=(10, 6))
plt.pie(crime_data2021, labels=crime_data2021.index, autopct='%1.1f%%', startangle=90, colors=sns.color_palette('Set3'))
plt.title('Crime type distribution in 2021')
plt.tight_layout()
plt.show()
```

* ####  To visually represent the distribution of different crime types against women in India for the year 2021.

* ####  Domestic violence and assault against women were the most reported crimes in 2021, making up over 60% of the total.

* #### women trafficking was the least reported crime in 2021, making up only 0.3% of the total.

### Domestic violence:
* ### Domestic violence is a significant issue, especially during the COVID-19 lockdowns,when calls to helplines increased. 
* ### For example, the National Commission for Women (NCW) received over 1,400 complaints of domestic violence in just the first month of the lockdown in 2020,

### Dowry Deaths:
* ### Although dowry deaths represent a smaller percentage in the pie chart, they remain a significant issue in rural areas and specific states like Bihar and Uttar Pradesh.

## Relationship Between Different Crime Types

```python
correlation_matrix = df[['Rape', 'Kidnapping and Abduction', 'Dowry Deaths', 'Assault against Women',
                         'Insult to Modesty of Women', 'Domestic Violence', 'Women Trafficking']].corr()
correlation_matrix
```

```python
plt.figure(figsize=(10, 6))
sns.heatmap(correlation_matrix, annot=True, cmap='RdPu')
plt.title('Correlation Matrix of Crimes Against Women (2001-2021)')
plt.show()
```

#### The correlation matrix shows that crimes like Rapes and assault against women are highly correlated, indicating they often occur together.
#### In contrast, women trafficking has the lowest correlation with other crimes, suggesting it occurs more independently.
